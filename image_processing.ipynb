```python
import numpy as np
from skimage import io # to read image
import matplotlib.pyplot as plt # to show (and save) image
import copy

def show(img):
    # ensure images are tight
    plt.axis('off')  # Turn off axis ticks and labels
    plt.tight_layout(pad=0)  # Remove padding around the imag
    plt.imshow(img)
 
img = io.imread('grumpycat.png') # read image, is stored as numpy-array in variable img
# Check image shape (height, width, color channels)
img.shape
```

```python
i = copy.deepcopy(img).astype('int16')

i = i*-1 + 255
show(i)
```

```python
i = copy.deepcopy(img)

for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        img[row, col, 1:3] = 0

i[:, :, 1:3] = 0
show(i)
```

```python
i = copy.deepcopy(img)

threshold = 128
for y in range(i.shape[0]):
    for x in range(i.shape[1]):
        pixel = i[y, x]
        i[y, x] = np.zeros(3, dtype=int) if np.average(pixel) < threshold else np.ones(3, dtype=int)*255

show(i)
```

```python
i = copy.deepcopy(img)

for y in range(i.shape[0]):
    for x in range(i.shape[1]):
        pixel = i[y, x]
        i[y, x] = np.average(pixel)

# i = np.apply_along_axis(np.max, axis=2, arr=i).repeat(3).reshape(i.shape).astype(int)
show(i)
```

```python
i = copy.deepcopy(img)

i[:,:,1:3] = 0  # for all columns and rows, zero out channels 1 & 2
show(i)
```

```python
i = copy.deepcopy(img)

show(i[::-1])
```

```python
i = copy.deepcopy(img)

downsample=12

i = i[::downsample,::downsample]
show(i)
```

```python
i = copy.deepcopy(img)
i[::10] = np.zeros((i.shape[1], 3), dtype=int)
i[:, ::10] = np.zeros(( 3), dtype=int)
show(i)
```
